# Logistic Regression - study

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
plt.style.use('seaborn-dark')
```

```python
# generate probabilities
x = np.arange(0,1,0.05) # consider these as probabilities of some event

odds = []
lOdds = []
def oddsRatio(p):
    o = p/(1-p)
    return o
  
def logOdds(x):
    lOdds = np.log(x)
    return lOdds
    
for i in x:
    val = oddsRatio(i)
    odds.append(val)
    
odds = np.array(odds)
max(odds)
len(odds)

for i in odds:
    val = logOdds(i)
    lOdds.append(val)
  
lOdds = np.array(lOdds)




```

```python
#probabilities

plt.figure(figsize = (18,20))
plt.plot(x, label = 'probabilities')
plt.title('Proababilities', fontsize = 30)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.ylim = (0,1)
plt.xlim = (0,20)
plt.grid('True')
```

```python
# odds
plt.figure(figsize = (18,20))
plt.plot(odds, label = 'odds')
plt.title('Odds of Probabilities', fontsize = 30)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlim=(0,20)
plt.ylim=(0,20)
plt.grid('True')

```

```python
# Probabilities vs Odds
plt.figure(figsize = (18,20))
plt.plot(x,odds, label = 'prob Vs odds')
plt.xlabel('Probabilities', fontsize = 20)
plt.ylabel('odds', fontsize = 20)
plt.title('Odds Vs Probabilities', fontsize = 30)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.legend()
plt.grid('True')

```

```python
# Log of odds
plt.figure(figsize = (18,20))
plt.plot(x,lOdds)
plt.xlabel('Probabilities', fontsize = 20)
plt.title(' Probabilities Vs Log of Odds', fontsize = 30)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.grid('True')
plt.plot(odds, label = 'odds')
plt.plot(lOdds, label = 'log of odds')
plt.legend()
min(lOdds)
```
