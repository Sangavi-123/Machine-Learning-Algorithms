# Sigmoid Function

Import dependencies

```python
import numpy as np
import pandas as pd
from random import randint
import matplotlib.pyplot as plt
```

### Hypothesis Function
A linear model is created. "hyp = 0.5 + 0.2x"

The parameters are:
theta0 = 0.5,
theta1 = 0.2

Outcomes of h(x) is calculated for 200 values in the range(-100,100)

```python
h_of_x = []
def hypothesis(x):
  
    "Calculates the hypothesis function . This function is linear"
    hyp = 0.5 + 0.2*(x)
    h_of_x.append(hyp)
  
    
# generate a range of input values
inputFeatures = [randint(-100,100) for i in range(200)] 
for i in inputFeatures:
    hypothesis(i)
  
```

```python
# Inverse logit or the sigmoid transformation
# e = 2.7182
sigmoidValues = [ ]

def sigmoid(x):
    "This function transforms or maps each x into a value in the range (0,1)"
    sig = 1/(1 + np.exp(-x))
    sigmoidValues.append(sig)
    
sigmoidValues.sort()
inputFeatures.sort()
for i in h_of_x:
    sigmoid(i)
    
plt.plot(inputFeatures)

sigmoidValues.sort()
plt.figure(figsize = (18,20))    
plt.plot(inputFeatures,sigmoidValues,linewidth = 7)
plt.axhline(0.5, color = 'green', alpha = 0.5,linestyle = ':')
plt.axvline(0,color = 'green', alpha = 0.5,linestyle = ':')
plt.axhline(0, color = 'black',alpha = 0.5, linestyle = ':')
plt.axhline(1, color = 'black',alpha = 0.5, linestyle = ':')
plt.title('Sigmoid transformation', fontsize = 30)

min(sigmoidValues)
max(sigmoidValues)
```
